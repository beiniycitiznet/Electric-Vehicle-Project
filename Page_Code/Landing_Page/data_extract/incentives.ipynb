```python
state=[]
title=[]
text=[]
type_=[]

for entry in response_json['result']:

    state.append(entry['state'])
    text.append(entry['text'])
    title.append(entry['title'])
    type_.append(entry['type'])  
```

```python
data=pd.DataFrame({'State':state, 'Title':title, 'Description':text, 'Type':type_})

data['State'].unique()
```

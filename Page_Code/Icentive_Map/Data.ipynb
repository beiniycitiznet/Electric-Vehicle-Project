```python
# Import modules
import pandas as pd
import json
import pymongo
```

```python
# Create a file path
files = "templates/incentives.csv"
```

```python
# Read CSV File
unorganized = pd.read_csv(files)
unorganized
```

```python
# df=unorganized.groupby(['State'],as_index=False).agg(lambda x : x.sum() if x.dtype=='float64' else ','.join(x))
```

```python
# Create a dataframe to count the type by state
total_count_df = unorganized.groupby(['State'],as_index=False)["Description"].count()
total_count_df = total_count_df.rename(columns={"Description": "Total_Type_Count"})
total_count_df.head()
```

```python
# Create a dataframe grouped by state and type to show the count of each type
type_count_df = unorganized.groupby(['State','Type'])["Description"].nunique()
type_count_df = type_count_df.reset_index().rename(columns={"Type":"Type_Count_Name", "Description": "Type_Count"})
type_count_df.head()
```

```python
# Add type name column and type count into a single column to add in json file

Specific_Count=[]
for index, row in type_count_df.iterrows():
    name = type_count_df["Type_Count_Name"][index]
    number = type_count_df["Type_Count"][index]
    Specific_Count.append(f"{name} : {number}")
    
# Specific_Count
    
type_count_df["Specific_Count"]=Specific_Count

type_count_df.drop(columns = ["Type_Count_Name", "Type_Count"], inplace=True)

type_count_df.head()
```

```python
# Create a function to use on a groupby on strings
def join_and_remove_duplicates(x: list, sep: str) -> list:
    deduped = []
    for i in x:
        if i not in deduped:
            deduped.append(i)
    return sep.join(deduped)
```

```python
# df=unorganized.groupby(['State'],as_index=False).agg(lambda x : x.sum() if x.dtype=='float64' else "|".join(set(x)))
# df.head(10)
```

```python
# Group the states together to place in to json file for the description
specific_count_df=type_count_df.groupby(['State'],as_index=False).agg(lambda x : x.sum() if x.dtype=='float64' else join_and_remove_duplicates(x, "<br>"))
specific_count_df.head(10)
```

```python
# Group the csv file by the state
df=unorganized.groupby(['State'],as_index=False).agg(lambda x : x.sum() if x.dtype=='float64' else join_and_remove_duplicates(x, "<hr>"))
df.head(10)
```

```python
# Create a list of states in abbreviation
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
          "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
# Create the list by full name
fullname = ['Alabama','Alaska','Arizona','Arkansas',
    'California','Colorado','Connecticut','Delaware',
    'District of Columbia','Florida','Georgia','Hawaii',
    'Idaho','Illinois','Indiana','Iowa','Kansas',
    'Kentucky','Louisiana','Maine','Maryland',
    'Massachusetts','Michigan','Minnesota','Mississippi',
    'Missouri','Montana','Nebraska','Nevada',
    'New Hampshire','New Jersey','New Mexico','New York',
    'North Carolina','North Dakota','Ohio','Oklahoma',
    'Oregon','Pennsylvania','Rhode Island','South Carolina',
    'South Dakota','Tennessee','Texas','Utah',
    'Vermont','Virginia','Washington','West Virginia',
    'Wisconsin','Wyoming']
```

```python
# Make the list in dataframe
us_state_abbrev = pd.DataFrame({"State": states, "Full_Name": fullname})
us_state_abbrev.head()
```

```python
# Merge the data frames to add full names of the state
new_incentive_df = pd.merge(left=df, right=us_state_abbrev, on="State")
new_incentive_df.head()
```

```python
# Merge the data frames togegher to get the count
new_incentive_df = pd.merge(left=new_incentive_df, right=total_count_df, on="State")
new_incentive_df.head()
```

```python
# Merge the dataframe for the description
new_incentive_df = pd.merge(left=new_incentive_df, right=specific_count_df, on="State")
new_incentive_df.head()
```

```python
# Open json file
with open("templates/us.json", "r") as read_file:
    data = json.load(read_file)
```

```python
# Read specific block in json 
z=data['features']
# z[0]['properties']
```

```python
# Add blocks into Json file
for info in z:
    for index,each in new_incentive_df.iterrows():
#         print(each["Full_Name"])
        if info['properties']["name"] == each["Full_Name"]:
            info['properties'].update({"Incentive_Title": each["Title"]})
            info['properties'].update({"Incentive_Description": each["Description"]})
            info['properties'].update({"Incentive_Type": each["Type"]})
            info['properties'].update({"Total_Type_Count": each["Total_Type_Count"]})
            info['properties'].update({"Specific_Count": each["Specific_Count"]})
            
            
```

```python
data['features'] = z
```

```python
# Write in the json file
with open('templates/refinedData.json', 'w') as outfile:
    json.dump(data, outfile)
```

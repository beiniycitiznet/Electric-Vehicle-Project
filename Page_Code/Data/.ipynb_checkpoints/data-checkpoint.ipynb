```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
from pprint import pprint
import us
```

```python
url='https://api.openchargemap.io/v3/poi/?output=json&countrycode=US&maxresults=32000&compact=true'
```

```python
charging_station_info = requests.get(url).json()
```

```python
len(charging_station_info)
```

```python
with open('data.json', 'w') as f:
    json.dump(charging_station_info, f)
```

```python
org_state_list=[]
for charging_station in charging_station_info:
    state_name=charging_station['AddressInfo']['StateOrProvince']
    if state_name !=None:
        org_state_list.append(state_name.upper())
        state_list=[x for x in org_state_list if len(x)==2]
```

```python
state_dict = {i:state_list.count(i) for i in state_list}
print(len(state_dict))
```

```python
state_dict
state_df=pd.DataFrame({"State":list(state_dict.keys()), "Number of Charging Stations":list(state_dict.values())})
state_df
```

```python
state_df['Number of Charging Stations']=state_df['Number of Charging Stations'].astype('int')
station_mean=state_df['Number of Charging Stations'].mean()
station_block=station_mean/3
station_median=state_df['Number of Charging Stations'].median()
station_median
```

```python
list_sum=0
station_block_list=[]
for i in range (1,7):
    station_block_list.append(int(station_block*i))
    list_sum+=station_block_list[i-1]
    
station_block_list.append(max(state_df['Number of Charging Stations'])-list_sum+10)
station_block_list
```

```python
column_name_list=['Number of Charging Stations 0 - 176','Number of Charging Stations 177 - 353', 'Number of Charging Stations  354 - 529', 'Number of Charging Stations 530 - 706', 'Number of Charging Stations 707 - 882', 'Number of Charging Stations 883 - 1059', 'Number of Charging Stations > 1059']
column_name_list
```

```python
for i, number in enumerate(station_block_list):
    state_df[f'{number}']=state_df.T.apply(lambda row:min(row.iloc[1]-(row.iloc[2:i+2]).sum(), number))
state_df=state_df.rename(columns={'176': column_name_list[0], '353': column_name_list[1], '529': column_name_list[2], '706': column_name_list[3], '882': column_name_list[4], '1059': column_name_list[5], '3526': column_name_list[6]})
state_df
```

```python
state_df=state_df.drop(['Number of Charging Stations'], axis=1)
state_df
```

```python
state_df.to_csv("station_state.csv",index=False)
```
